```python
import pandas as pd
import numpy as np
import os
```

```python
directory = "../DGreen_test_papers-20230726T075419Z-001/DGreen_test_papers/"
```

```python
refined_ = os.path.join(directory, "refined_paper_details.txt")
f = open(refined_)
f_content = f.readlines()
print(len(f_content))
```

```python
# Compute indixes for determining right key for title, author, ...
# create list of titles, authors, etc.
len_t = len("Title: ")
len_a = len("Author: ")
len_p = len("Publisher: ")
len_d = len("Date of publication: ")
titles = []
authors = []
publishers = []
dates_p = []
for line in f_content:
    end_index = len(line) - 1
    if line.startswith("Title"):
        titles.append(line[len_t:end_index])
    elif line.startswith("Author"):
        authors.append(line[len_a:end_index])
    elif line.startswith("Publisher"):
        publishers.append(line[len_p:end_index])
    elif line.startswith("Date"):
        dates_p.append(line[len_d:end_index])
```

```python
# Creating a dataframe
data = {"Title_pred": titles, "Author_pred": authors, "Publisher_pred": publishers, "Date_published_pred": dates_p}
df = pd.DataFrame(data)
```

```python
# loading csv file with paper details entered manually
df_papers = pd.read_csv('../DGreen_test_papers-20230726T075419Z-001/DGreen_test_papers/Papers_details.csv')
df_papers.head()
```

```python
df
```

```python
# Preprocessing values in dataset (df) with chatGPT extracted details
# Fill N/A, Not Specified, Unkown with NaNs
to_replace = ["N/A", "Not Specified.", "Not Specified", "Unknown", "Unknown.", "Unknow", "N/A."]
for s in to_replace:
    df.replace(s, np.nan, inplace=True)              
```

```python
merged_authors["Author_pred"] = merged_authors["Author_pred"].str.replace(" and", ",")
merged_authors["Author_pred"] = merged_authors["Author_pred"].str.replace(";", ",")
```

```python
merged_authors.head()
```

```python
merged_authors["Author"] = [ ",".join(author.split(", ").sort()) for author in merged_authors["Author"]]
```

```python
new_author = []
for author in merged_authors["Author"]:
    if type(author) == str:
        split_au = author.split(", ")
        split_au.sort()
        split_au = ",".join(split_au)
        new_author.append(split_au)
    else:
        new_author.append(author)
merged_authors["Author"] = new_author
```

```python
new_author_pred = []
for author in merged_authors["Author_pred"]:
    if type(author) == str:
        split_au = author.split(", ")
        split_au.sort()
        split_au = ",".join(split_au)
        new_author_pred.append(split_au)
    else:
        new_author_pred.append(author)
merged_authors["Author_pred"] = new_author_pred
```

```python
merged_authors
```

```python
group_1 = list(np.zeros(len(merged_authors), dtype=int))
group_2 = list(np.zeros(len(merged_authors), dtype=int))
group_3 = list(np.zeros(len(merged_authors), dtype=int))
group_4 = list(np.zeros(len(merged_authors), dtype=int))
index = -1
for author, author_pred in zip(merged_authors["Author"], merged_authors["Author_pred"]):
    index += 1
    if type(author) == float and type(author_pred) == float:
        group_4[index] = 1
        continue
    if type(author) == float and type(author_pred) != float:
        group_2[index] = 1
        continue
    if type(author) != float and type(author_pred) == float:
        group_1[index] = 1
        continue
    print(author)  
    authors = author.split(",")
    authors_pred = author_pred.split(",")
    right = False
    wrong = False
    count_all = 0
    for au_pred in authors_pred:
        if au_pred in authors:
            right = True
            count_all += 1
        else:
            wrong = True
        
    if right and wrong:
        group_2[index] = 1
        group_3[index] = 1
    
    if right and wrong == False:
        group_3[index] = 1
    
    if right and wrong == False and len(authors) == len(authors_pred):
        group_4[index] =  1
        
    if right and wrong and len(authors) == count_all:
        group_4[index] =  1
        group_2[index] =  1
    
    if right == False:
        group_2[index] = 1
```

```python
merged_authors["None identified"] = group_1
```

```python
merged_authors["Wrong author"] = group_2
```

```python
merged_authors["Some authors"] = group_3
```

```python
merged_authors["All authors"] = group_4
```
